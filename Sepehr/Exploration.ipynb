```python
from scipy.io import loadmat
import pandas as pd
```

```python
def matdata_to_pandasdf(subject_id, kind = 'self'):
    """
    Loads a matlab file and returns a pandas dataframe.
    """
    # mat = loadmat()

    if(kind == 'self'):
        d = loadmat('./play for self/subject_'+ str(subject_id) +'/data.mat')
    else:
        d = loadmat('./play for other/subject_'+ str(subject_id) +'/data.mat')
    
    mdata = d['data']
    mdtype = mdata.dtype 
    ndata = {n: mdata[n][0, 0] for n in mdtype.names}
        
    dataframe_experiment = {}
    if(subject_id == 16 and kind == 'other'):
        # print(list(ndata['rewardProbs']))
        dataframe_experiment['rewardProbs'] = list(ndata['rewardProbs'])[:-1]
    else:
        dataframe_experiment['rewardProbs'] = list(ndata['rewardProbs'])

    # print(len(dataframe_experiment['rewardProbs']))
    dataframe_experiment['state1'] = ndata['state1'][0]
    # print(len(dataframe_experiment['state1']))
    dataframe_experiment['choice'] = ndata['choice'][0]
    # print(len(dataframe_experiment['choice']))
    dataframe_experiment['resp_time_1'] = ndata['resp_time_1'][0]
    # print(len(dataframe_experiment['resp_time_1']))
    dataframe_experiment['pic_place'] = ndata['pic_place'][0]
    # print(len(dataframe_experiment['pic_place']))
    dataframe_experiment['reward'] = ndata['reward'][0]
    # print(len(dataframe_experiment['reward']))

    dataframe_experiment = pd.DataFrame(dataframe_experiment)

    dataframe_experiment['name'] = ndata['name'][0]
    dataframe_experiment['last_name'] = ndata['last_name'][0]
    # dataframe_experiment['edu'] = ndata['edu'][0]
    # dataframe_experiment['age'] = ndata['age'][0][0]
    dataframe_experiment.rewardProbs = dataframe_experiment.rewardProbs.map(lambda x: x[0])

    dataframe_experiment['rewardProbMAX'] = dataframe_experiment.rewardProbs.map(lambda x: max(x))
    dataframe_experiment['rewardProbMIN'] = dataframe_experiment.rewardProbs.map(lambda x: min(x))

    dataframe_experiment['ID'] = subject_id


    dataframe_experiment = dataframe_experiment.drop(['rewardProbs'], axis=1)
    dataframe_experiment = dataframe_experiment[['ID','name',
       'last_name', 'state1', 'choice', 'resp_time_1', 'pic_place', 'reward', 'rewardProbMAX', 'rewardProbMIN'  ]]

    return dataframe_experiment
```

```python
final_df_self = pd.DataFrame()

for i in range(1,39):
    df = matdata_to_pandasdf(i, 'self')
    # print(i)
    final_df_self = final_df_self.append(df)
final_df_self
```
